```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd
```

```python
import os
path_root = os.getcwd()
path_raw_data = os.path.join(path_root,'raw_data')
path_chromedriver=os.path.join(path_root,'chromedriver.exe')
```

```python
chrome_options = Options()
chrome_options.add_argument("--incognito")
chrome_options.add_argument("--start-maximized")
```

<h1> Crawl thegioididong.com </h1>

```python
driver = webdriver.Chrome(options=chrome_options, executable_path=path_chromedriver)

url_1 = 'https://www.thegioididong.com/dtdd/samsung-galaxy-a50/danh-gia'
url_2 = 'https://www.thegioididong.com/dtdd/samsung-galaxy-a50-128gb/danh-gia'

driver.get(url_1)
time.sleep(5)
```

```python
i=0
data = []
driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
num_page=driver.find_elements_by_xpath('//div[@class="pgrc"]/div/a')[-2].text
while (i<int(num_page)):
    list_star=[]
    names = driver.find_elements_by_xpath("//ul[@class='ratingLst']/li[@class='par ']/div[@class='rh']/span")
    comments = driver.find_elements_by_xpath("//ul[@class='ratingLst']/li[@class='par ']/div[@class='rc']/p/i")
    rows = driver.find_elements_by_xpath("//ul[@class='ratingLst']/li[@class='par ']/div[@class='rc']")
    for row in rows:
        list_star.append(len(row.find_elements_by_xpath("./p/span/i[@class='iconcom-txtstar']")))
    for n,c,s in zip(names,comments,list_star):
        data.append([n.text,c.text,s])
    driver.find_elements_by_xpath('//div[@class="pgrc"]/div/a')[-1].click()
    time.sleep(1)
    i=i+1
driver.close()
raw_data=pd.DataFrame(data,columns=['name','comment','star'])
```

<h1> Crawl tiki.vn </h1>

```python
driver = webdriver.Chrome(options=chrome_options, executable_path=path_chromedriver)
url_1 = 'https://tiki.vn/dien-thoai-samsung-galaxy-a50-64gb-4gb-hang-chinh-hang-p11508168.html'
url_2 = 'https://tiki.vn/dien-thoai-samsung-galaxy-a50-64gb-4gb-hang-chinh-hang-da-kich-hoat-bao-hanh-dien-tu-p12206668.html'
driver.get(url_2)
time.sleep(8)
```

```python
btn_cancel = driver.find_element_by_xpath('//button[@id="onesignal-popover-cancel-button"]')
btn_cancel.click()
time.sleep(2)
driver.execute_script("window.scrollTo(0, 4000);")
time.sleep(2)
```

```python
data = []
while (True):
    list_star=[]
    names = driver.find_elements_by_xpath('//div[@class="user-info"]/p[@class="username"]')
    comments = driver.find_elements_by_xpath('//div[@class="information"]/p[@class="content"]')
    rows = driver.find_elements_by_xpath('//div[@class="title"]/p')
    for row in rows:
        list_star.append(len(row.find_elements_by_xpath('./i[@class="icomoon icomoon-star"]')))
    for n,c,s in zip(names,comments,list_star):
        data.append([n.text,c.text,s])
    try:
        driver.find_elements_by_xpath('//ul[@class="Navigation__Wrapper-dgcpmq-0 kiFxlu"]/li/a[@class="btn next"]')[0].click()
        time.sleep(1)
    except:
        break
driver.close()
raw_data = pd.DataFrame(data,columns=['name','comment','star'])
```

<h1> Concate file </h1>

```python
tgdd_64 = pd.read_csv(path_raw_data+'/'+'tgdd_ss_a50_64.csv')
tgdd_128 = pd.read_csv(path_raw_data+'/'+'tgdd_ss_a50_128.csv')
tiki_206 = pd.read_csv(path_raw_data+'/'+'tiki_ss_a50_206.csv')
tiki_245 = pd.read_csv(path_raw_data+'/'+'tiki_ss_a50_246.csv')
```

```python
raw_data_ = pd.concat([tgdd_64,tgdd_128,tiki_206,tiki_245])
```

```python
raw_data_.drop_duplicates(subset = 'comment',keep = 'first', inplace = True)
raw_data_.name = [i.title() for i in raw_data_.name]
```
